```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns 
```

```python
df = pd.read_csv(r"D:\CODECRAFT_DS_01\dataset\API_SP.POP.TOTL_DS2_en_csv_v2_40826.csv", skiprows=4)
```

```python
df.head()
```

```python
df.shape
```

```python
df = df[['Country Name', '2022']]
df.head()
```

```python
df = df.dropna()
df.shape
```

```python
top10 = df.sort_values(by='2022', ascending=False).head(10)

plt.figure(figsize=(12,6))
sns.barplot(x='Country Name', y='2022', data=top10)

plt.xticks(rotation=45)
plt.title("Top 10 Countries by Population (2022)")
plt.xlabel("Country")
plt.ylabel("Population")
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.hist(df['2022'], bins=20)

plt.title("Distribution of Population (2022)")
plt.xlabel("Population")
plt.ylabel("Frequency")
plt.show()
```

## Conclusion

- India and China have the highest population.
- Most countries fall in lower population range.
- Distribution is right-skewed.
